```python
from tkinter import *
from tkcalendar import *
from datetime import datetime
import datetime

root = Tk()
root.title("Doctor Appointment System")
root.geometry('400x300')

# Login screen
def login():
    username = username_entry.get()
    password = password_entry.get()
    if username == "doctor" and password == "password":
        main_menu()
    else:
        error_label = Label(root, text="Invalid login credentials")
        error_label.place(x=140,y=270)

heading=Label(root,text="Doctor Appointment System",font=('arial',20,'bold'))
heading.place(x=10,y=5)

style1 = Label(root,bg="#248aa2",height=1,width=50)
style1.place(x=0,y=50)
style1 = Label(root,bg="#248aa2",height=1,width=30)
style1.place(x=300,y=50)

frame1 = LabelFrame(root,text="Login",width=200,height=200,font=('verdana',10,'bold'))
frame1.place(x=100,y=80)

username_label = Label(root, text="Username:")
username_label.place(x=170,y=120)
username_entry = Entry(root)
username_entry.place(x=140,y=140)
password_label = Label(root, text="Password:")
password_label.place(x=170,y=170)
password_entry = Entry(root, show="*")
password_entry.place(x=140,y=190)
login_button = Button(root, text="Login", command=login)
login_button.place(x=180,y=230)


# Main menu
def main_menu():
    top= Toplevel(root)
    top.geometry("400x300")
    top.title("Slot Booking")
    menu_label = Label(top, text="Select a date for your appointment:",font=('arial',10,'bold'))
    menu_label.place(x=0,y=0)

    # Calendar widget
    cal = Calendar(top, selectmode='day', year=2023, month=4, day=29)
    cal.place(x=50,y=50)
    
    style2 = Label(top,bg="#248aa2",height=1,width=50)
    style2.place(x=0,y=22)
    style2 = Label(top,bg="#248aa2",height=1,width=30)
    style2.place(x=300,y=22)

    def register_patient():
        registration_window = Toplevel(root)
        registration_window.geometry("400x300")
        registration_window.title("Patient Registration Form")
        r_label= Label( registration_window, text="Register Yourself",font=('arial',10,'bold'))
        r_label.place(x=0,y=0)
        
        style3 = Label(registration_window,bg="#248aa2",height=1,width=50)
        style3.place(x=0,y=30)
        style3 = Label(registration_window,bg="#248aa2",height=1,width=30)
        style3.place(x=300,y=30)

        # Registration form
        name_label = Label(registration_window, text="Name:")
        name_label.place(x=0,y=60)
        name_entry = Entry(registration_window,width=40)
        name_entry.place(x=120,y=60)
        dob_label = Label(registration_window, text="Date of Birth:")
        dob_label.place(x=0,y=100)
        dob_entry = Entry(registration_window,width=40)
        dob_entry.place(x=120,y=100)
        phone_label = Label(registration_window, text="Phone Number:")
        phone_label.place(x=0,y=140)
        phone_entry = Entry(registration_window,width=40)
        phone_entry.place(x=120,y=140)
        email_label = Label(registration_window, text="Email:")
        email_label.place(x=0,y=180)
        email_entry = Entry(registration_window,width=40)
        email_entry.place(x=120,y=180)
        medical_label = Label(registration_window, text="Medical History:")
        medical_label.place(x=0,y=220)
        medical_entry = Entry(registration_window,width=40)
        medical_entry.place(x=120,y=220)

        

        def confirm_registration():
            name = name_entry.get()
            dob = dob_entry.get()
            phone = phone_entry.get()
            email = email_entry.get()
            medical = medical_entry.get()
            
            confirmation_window = Toplevel(root)
            confirmation_window.geometry("400x300")
            confirmation_window.title("Appointment Confirmation")

            # Confirmation screen
            confirmation_label = Label(confirmation_window, text="Appointment Confirmed",font=('arial',10,'bold'))
            confirmation_label.place(x=0,y=0)
            
            style3 = Label(confirmation_window,bg="#248aa2",height=1,width=50)
            style3.place(x=0,y=30)
            style3 = Label(confirmation_window,bg="#248aa2",height=1,width=30)
            style3.place(x=300,y=30)
            
            name_label = Label(confirmation_window, text="Name: " + name)
            name_label.place(x=100,y=60)
            dob_label = Label(confirmation_window, text="Date of Birth: " + dob)
            dob_label.place(x=100,y=90)
            phone_label = Label(confirmation_window, text="Phone Number: " + phone)
            phone_label.place(x=100,y=120)
            email_label = Label(confirmation_window, text="Email: " + email)
            email_label.place(x=100,y=150)
            medical_label = Label(confirmation_window, text="Medical History: " + medical)
            medical_label.place(x=100,y=180)
            date_label = Label(confirmation_window, text="Appointment Date: " + str(cal.selection_get()))
            date_label.place(x=100,y=210)
            doctor_label = Label(confirmation_window, text="Doctor: Dr. Smith")
            doctor_label.place(x=100,y=240)
            

        confirm_button = Button(registration_window, text="Confirm Registration", command=confirm_registration)
        confirm_button.place(x=140,y=250)

    register_button = Button(top, text="Register Patient", command=register_patient)
    register_button.place(x=130,y=250)

    def confirm_appointment():
        date = cal.selection_get()
        confirmation_label = Label(root, text="Your appointment has been scheduled for " + str(date))
        confirmation_label.pack()
        # Disable the selected date in the calendar
        cal.calevent
    
    def set_appointments():
        for date in appointments:
            year, month, day = map(int, date.split('-'))
            cal_date = cal._date(year, month, day)
            if appointments[date]:
                cal.calevent_create(cal_date, appointments[date], 'booked', 'red')
            else:
                cal.calevent_create(cal_date, appointments[date], 'available', 'green')

root.mainloop()
```
